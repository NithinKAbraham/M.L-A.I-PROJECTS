## <span style="text-decoration: underline; color: violet;">AIRFARE INFLUENCERS</span>

##### Airline fare prediction is influenced by a multitude of factors that can affect ticket prices.Understanding these factors is essential for 

##### accurate fare  predictions.Here are some of key elements that play a role:

###### <span style="text-decoration: underline; color: red;">1.Booking Time:</span>

######           The timing of your ticket  purchase is crucial ,Fares tend to be higher when booking closer to the departure date,
###### especially during peak travel seasons

###### <span style="text-decoration: underline; color: red;">2.Day Of The Week :</span>

###### Airline often adjust prices basesd on the day of week.Mid-week days like Tuesday and Wednesday may have lower fares compared to 

###### weekends


###### <span style="text-decoration: underline; color: red;">3. Time Of Day :</span>

######  The time of day when you book can impact prices.Some experts suggest that booking flights during non-peak hours can lead to 
######  better deals

###### <span style="text-decoration: underline; color: red;">4.Season And Demand :</span>

###### Traveling during peak seasons or holidays generally results in higher fares due to increased demand.Conversely, off-peak times may offer 

###### more affordable options

###### <span style="text-decoration: underline; color: red;">5. Flight Route :</span>

######   The specific route you choose can significantly affect the fare. Direct flights tend to be pricier than connecting flights, and different airlines 

######   may have varying pricing strategies for the same route.

###### <span style="text-decoration: underline; color: red;">6.Airline And Class of Service : </span>

######  Different airlines price their tickets differently.Addionally,the class of service (ECONOMY,BUSINESS,FIRST CLASS) also impact the 
######   cost of a tickets.

###### <span style="text-decoration: underline; color: red;">7.Fuel Prices :</span>

######  Fluctuations in oil and jet fuel prices can impact airline operating costs,which may be reflected in ticket prices.

###### <span style="text-decoration: underline; color: red;">8.Competition :</span>

###### The level of competition on a specific route can influence prices. More competition often results in lower fares.

###### <span style="text-decoration: underline; color: red">9.Flexibility :</span>

###### Being flexible with your travel dates,airports,and even destinations can lead to cost savings.

###### <span style="text-decoration: underline; color: red;">10.Economic Factors :</span>

###### The overall state of the economy can influence travel demand and pricing.

<span style="text-decoration: underline; color: blue;">G. FIXING TIME-ZONES</span>

```python
def categorize_departure_time(time_str):
    hour=int(time_str.split(':')[0])
    
    if 5<=hour<12:
        return "Morning"
    elif 12<=hour<16:
        return "Afternoon"
    elif 16<=hour<20:
        return "Evening"
    elif 20<=hour<24 or 0<=hour<5:
        return "Night"
    else:
        return "unknown"
    
df1['Departure Time']=df1['Departure Time'].apply(categorize_departure_time)
        
```

```python
def categorize_arrival_time(time_str):
    hour=int(time_str.split(':')[0])
    
    if 5<=hour<12:
        return "Morning"
    elif 12<=hour<16:
        return "Afternoon"
    elif 16<=hour<20:
        return "Evening"
    elif 20<=hour<24 or 0<=hour<5:
        return "Night"
    else:
        return "unknown"
    
df1['Arrival_Time']=df1['Arrival Time'].apply(categorize_arrival_time)
        
```

### <span style="text-decoration: underline; color: red;">1. ENCODING</span>

```python
import hashlib
```

```python
def feature_hashing(df1, columns, num_buckets):
    for col in columns:
        df1[col + '1'] = df1[col].apply(lambda x: int(hashlib.sha256(x.encode('utf-8')).hexdigest(), 16) % num_buckets)
    return df1

num_buckets = 10 
columns_to_hash = ['Airline', 'Flight']

# Perform feature hashing
df1 = feature_hashing(df1, columns_to_hash, num_buckets)
df1
```
